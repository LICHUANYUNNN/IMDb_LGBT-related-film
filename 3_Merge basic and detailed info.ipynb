```python
import pandas as pd 
df1 = pd.read_csv('IMDb basicinfo_B.csv', encoding = 'utf8',index_col=0)
df2 = pd.read_csv('IMDb detailinfo_B1.csv', encoding = 'utf8',index_col=0)
```

```python
df3 = df2.drop(['name'],axis=1)
```

```python
df4 = df1.join(df3, on=None, how='left', lsuffix='', rsuffix='',sort=False)
```

```python
df5 = df4[['name','year','genre','rate','voter','director','star','country','company','budget','gross']]
```
