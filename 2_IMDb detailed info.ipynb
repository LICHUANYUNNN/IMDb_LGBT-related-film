```python
import requests 
import bs4 as bs 
import pandas as pd 
import csv 
import urllib.request
from time import sleep
from random import randint
from time import time
```

```python
url = u'https://www.imdb.com'
def start_request(q):
    r = requests.get(url+q)
    print(url+q)
    return r.text
```

```python
def parse(text):
    soup = bs.BeautifulSoup(text, features='html.parser') 
    container1 = soup.find('div', attrs={'class': 'title-overview'})
    container2 = soup.find('div',attrs={'class':'titleReviewBarItem titleReviewbarItemBorder'})
    container3 = soup.find('div',attrs={'id':'main_bottom'})
    mydict = {}
    star = []
    country = []
    company = []
    try: 
        mydict['name'] = container1.find('h1').text 
        
        try:  
            mydict['rate'] = container1.find(itemprop = 'ratingValue').text
            mydict['voter'] = container1.find(itemprop = 'ratingCount').text
        except:
            mydict['rate'] = '0'
            mydict['voter'] = '0'
        for h4 in soup.find_all('h4'):
            if "Director:" in h4:
                mydict['director'] = h4.next_sibling.next_element.text.strip()
            if "Stars:" in h4:
                h = h4
                while h:
                    try:
                        h = h.next_sibling.next_element
                        star.append(h.text.strip() + ",")
                        mydict['star'] = "".join(star)
                    except Exception as e:
                        break
            if "Country:" in h4:
                h = h4
                while h:
                    try:
                        h = h.next_sibling.next_element
                        country.append(h.text.strip().replace("|",",")) 
                        mydict['country'] = "".join(country) 
                    except Exception as e:
                        break
            if "Production Co:" in h4:
                h = h4
                while h:
                    try:
                        h = h.next_sibling.next_element
                        company.append(h.text.strip() + ",")
                        mydict['company'] = "".join(company)
                    except Exception as e:
                        break
            if "Budget:" in h4:
                mydict['budget'] = h4.next_sibling.strip()
            if "Gross USA:" in h4:
                mydict['gross'] = h4.next_sibling.strip()
        result_list.append(mydict)
    except:
        print('not___________________________________________________found')
        mydict['name'] = 'null'
        mydict['rate'] = 'null'
        mydict['voter'] = 'null'
        mydict['director'] = 'null'
        mydict['star'] = 'null'
        mydict['country'] = 'null'
        mydict['company'] = 'null'
        mydict['budget'] = 'null'
        mydict['gross'] = 'null'
        result_list.append(mydict)
```

```python
def main(req):
    t = start_request(req)
    parse(t)
```

```python
def read_csv(filename):
    with open(filename) as f:    
        reader = csv.DictReader(f)    
        for row in reader:         
            aaa = row['id']  
            main(aaa)  
            sleep(randint(1,3))
```

```python
result_list=[]
read_csv('IMDb movieid_G.csv') 
```

```python
df = pd.DataFrame(result_list)
df.to_csv('IMDb detailinfo_G1.csv')
```
