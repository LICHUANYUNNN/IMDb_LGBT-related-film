```python
#raw dataset
df = df.drop(['star'],axis=1)
df = df.drop(['company'],axis=1)
df = df[['name','year','genre','rate','voter','director','country','gross','budget']]
df.head(15)
```

```python
#cleaned dataset
df['year'] = df['year'].str.extract(r'(\d{4})', expand=False)#year: extract the digits
df['voter'] = df['voter'].str.replace(",","")                #voter: remove thousand separators
df['voter'] = pd.to_numeric(df['voter'])

df['gross']=df['gross'].str.replace(",","")                  #gross: remove punctuations
df['gross']=df['gross'].str.replace("$","")
df['gross'] = pd.to_numeric(df['gross'])

df['country']=df['country'].fillna("None")                   #country: the first one
country1=[]
for i in df['country']:
    comma = i.find(',')
    if comma > 0:
        country1.append(i[0:comma])
    else:
        country1.append(i)
df['country']=country1
df = df[['name','year','genre','rate','voter','director','country','gross','budget']]
```

```python
#select films with the most voters (top 5) in Germany, France, Italy and USA
ge_df = df[df['country'] == 'Germany'].sort_values(by=['voter'], ascending=False).head(5)
fr_df = df[df['country'] == 'France'].sort_values(by=['voter'], ascending=False).head(5)
it_df = df[df['country'] == 'Italy'].sort_values(by=['voter'], ascending=False).head(5)
us_df = df[df['country'] == 'USA'].sort_values(by=['voter'], ascending=False).head(5)
df1 = ge_df.append(fr_df, ignore_index=True)
df2 = df1.append(it_df, ignore_index=True)
df3 = df2.append(us_df, ignore_index=True)
df3['director']=df3['director'].fillna("None")
df4 = df3.dropna()
df4
```

```python
#the exchange rate between USD and EURO is 0.89 (Nov 27, 2018), so here we choose to ignore the difference. 
df4['budget']=df4['budget'].str.replace(",","")
df4['budget']=df4['budget'].str.replace("$","")
df4['budget']=df4['budget'].str.replace("€","")
df4['budget'] = pd.to_numeric(df4['budget'])
df4['budget'] = df4['budget'].div(1000000)
df4['gross'] = df4['gross'].div(1000000)
df4
```
